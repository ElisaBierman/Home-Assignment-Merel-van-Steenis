Visualizing brain activation data on top of anatomical scan data to visualize fMRI brain activation for working memory.

```python
#importing the nibabel package
import nibabel as nib

# loading the MRI data into VR code
MRIdata = nib.load("anatomical.nii.gz")
#print(MRIdata)
# loading the Functional data of WM into VR code
Functional_data = nib.load("working memory_uniformity-test_z_FDR_0.01.nii.gz")
#print(Functional_data)

# plot the functional data on top of the anatomical data
from nilearn import plotting

plotting.plot_stat_map(
    Functional_data,  # plotting the functional data
    bg_img=MRIdata,  # on top of the MRI data (=background image)
    display_mode="ortho",  # axial, coronal, sagittal slices
    cut_coords=(-0.5, 1, 2),  # setting the cut coordinates
    threshold=0.1,  # setting the voxel treshold to 0.1
    cmap="hot",  # changing the color map
    title="Working Memory",  # adding a title
)

plotting.show()
```

This figure highlights voxels (3D pixels of an MRI image) whose activity has been linked to Working Memory in 332 neuroimaging studies. Data were downloaded from Neurosynth.org. XYZ coordinates reflect the 3D position of the crosshair in brain space. The color indicates the strength of association with the concept.

Plotting a histogram of the positive 1D voxel data of Working Memory fMRI data  

```python
# printing the functional data of Working Memory
#print(Functional_data)
# installing numpy
import numpy as np

# importing the matplotlib package for plotting a histogram
import matplotlib.pyplot as plt

# check the type of functional data
#type(Functional_data)
# changing the data of functional data to 'Numpy arrays'
Functionaldata = Functional_data.get_fdata()
#type(Functionaldata)
# Select only positive values of the functional data
Functionaldata = Functionaldata[Functionaldata > 0]
# Plot histogram
plt.figure(figsize=(8, 5))  # plotting a figure of size 8 by 5
plt.hist(
    Functionaldata.flatten(), bins=50, edgecolor="black"
)  # the .flatten() function changes the 3D MRI data to 1D data !!!
plt.title("Histogram of Activation Values for Working Memory fMRI")  # adding a title
plt.xlabel("Voxel intensity")  # adding an label to the x-axis
plt.ylabel("Number of Voxels")  # adding a label to the y-axis
plt.show()
```
